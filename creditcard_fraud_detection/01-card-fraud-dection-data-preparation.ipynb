Description: This is the first part of the project for detecting credit card fraud. It uses the ULB credit card fraud dataset (https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud) which contains anonymized 
credit card transaction details with the amount and a two class output of 0 (legitimate transaction) and 1 (fraudulent transaction)

```python
#import importan libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
df = pd.read_csv("/kaggle/input/datasets/andreasloizias/creditcardfraud/creditcard.csv")
df.head()
```

```python
df.shape
```

```python
#check for any null values
df.isnull().sum()
```

```python
df["Class"].value_counts()
#we see there are only 492 cases of fraud and 284315 cases of non-frau transactions
```

```python
#check percentage of fraud vs non-fraud cases
df["Class"].value_counts(normalize=True)
```

```python
# we see from above that the class is heavily imbalanced
# we plot the imbalance to visualize it
counts = df["Class"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(["Normal", "Fraud"], counts.values)

plt.title("Transaction Class Distribution")
plt.ylabel("Number of Transactions")

plt.show()
```

```python
# check the amounts
df["Amount"].describe()
```

```python
#visualize the amounts into bins
plt.figure(figsize=(8,4))
plt.hist(df["Amount"], bins=100)

plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
#compare the fraud vs normal transactions amounts
fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

plt.figure(figsize=(8,4))

plt.hist(normal["Amount"], bins=10, alpha=0.5, label="Normal")
plt.hist(fraud["Amount"], bins=10, alpha=0.5, label="Fraud")

plt.legend()
plt.title("Transaction Amount: Fraud vs Normal")

plt.show()
```

```python
#split DF to features and target
X = df.drop("Class", axis=1)
y = df["Class"]

print(X.shape)
print(y.shape)
```

```python
#split to training and test splitting 80:20% train:test
#shuffle using random seet 22
from sklearn.model_selection import train_test_split

X_train_full, X_test, y_train_full, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=22
)
```

```python
#further split the original training set to actual training set and evaluation set
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full,
    y_train_full,
    test_size=0.2,
    stratify=y_train_full,
    random_state=42
)
```

```python
#scale the features in case we use a NN since they perform better when the features are scaled
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
#save the sets to be used in the next step
np.save("X_train.npy", X_train)
np.save("X_val.npy", X_val)
np.save("X_test.npy", X_test)

np.save("y_train.npy", y_train)
np.save("y_val.npy", y_val)
np.save("y_test.npy", y_test)
```

```python
#save the scaler to be used later
import joblib
joblib.dump(scaler, "scaler.pkl")
```
